# Preparetion

```python
import numpy as np
import marsilea as ma
import marsilea.plotter as mp
```

# 1. How to arrange legends

Try to arrange the legends in order of anno2, main, anno1, from left to right

```python
data = np.random.randn(10, 10)

h = ma.Heatmap(data, label="main")
h.add_right(mp.ColorMesh(data[0], cmap="Purples", label="anno1"), size=0.5, pad=0.1)
h.add_right(mp.Colors(np.random.choice([0, 1], 10), label="anno2"), size=0.5, pad=0.1)
h.add_legends()
h.render()
```

# 2. How to label a plot

Try to place the label at the top.

```python
data = np.random.randn(10, 10)

h = ma.Heatmap(data, label="main")
h.add_right(mp.ColorMesh(data[0], cmap="Purples", label="anno1"), size=0.5, pad=0.1)
h.add_right(mp.Colors(np.random.choice([0, 1], 10), label="anno2"), size=0.5, pad=0.1)
h.render()
```

# 4. How to apply the dendrogram of a cross-layout to another

```python
data1 = np.random.randn(10, 10)
data2 = np.random.randint(0, 10, (10, 10))

h1 = ma.Heatmap(data1)
h1.add_right(mp.Labels(np.arange(10)))
h1.add_dendrogram("left")

h2 = ma.Heatmap(data2)
h2.add_left(mp.Labels(np.arange(10)))
h2.add_dendrogram("right")

(h1 + 0.5 + h2).render()
```

# 5. How to label the chunk of data after grouping

Can you label the 1 and 2 with a big title instead of labeling each row?

```python
data = np.random.randn(6, 6)

h = ma.Heatmap(data, label="main")
h.group_rows([1, 2, 1, 2, 2, 2])
h.add_right(mp.Labels([1, 2, 1, 2, 2, 2]))
h.render()
```

# 6. How to color the dendrogram

Can you make the base dendrogram and the meta dendrogram different colors

```python
data = np.random.randn(10, 10)

h = ma.Heatmap(data, label="main")
h.cut_cols([2, 6])
h.add_dendrogram("top")
h.render()
```
